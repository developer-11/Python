```python
import os
import requests

# Create a folder to save images
folder = "images"
if not os.path.exists(folder):
    os.makedirs(folder)

# Example image source (replace with your own links)
# Here I'm assuming you have image URLs in a sequence like image1.jpg, image2.jpg ...
base_url = "https://cdn.pixabay.com//images/image"

# Download 1000 images
for i in range(1, 10):
    url = f"{base_url}{i}.jpg"   # example: https://example.com/images/image1.jpg
    filename = os.path.join(folder, f"image{i}.jpg")

    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            with open(filename, "wb") as f:
                f.write(response.content)
            print(f"Downloaded: {filename}")
        else:
            print(f"Failed: {url} (Status {response.status_code})")
    except Exception as e:
        print(f"Error downloading {url} → {e}")

print("✅ Download complete!")
```

```python
#integer

print(7/3) # division
print(7//3) # floor division

#floating

print(0.1+0.2) # not precised value because float stored in binary

# to get a precised value we need to import Decimal from decimal library
from decimal import Decimal
print(Decimal('0.1')+Decimal('0.2'))

#Math Module

import math
print(math.sqrt(4))
print(math.factorial(5))
print(round(math.pi,2))
print(math.e)
print(math.log(10))
print(math.log10(10))
print(math.log(1))

#Random Module

import random
print(random.randint(1,100)) # it will generate random integer in the given range
print(random.random()) # it will generate random floating number in the range [0,1)
print(random.choice([1,2,3,4,10,22,34,45])) # it will choose any one random number from the list
print(random.choices([1,2,3,4,10,22,34,45],k=3)) # it will choose any "k" random number from the list
```

```python
#casting

# 1. Implicit Casting - data type converts automatically according to the use

x=10
y=2.5

z=x+y
print(z)
print(type(x),type(y),type(z))

# 2. Explicit Casting - data type converts by using built-in functions

# list ↔ tuple
print(list((1, 2, 3)))
print(tuple([4, 5, 6]))

# list ↔ set
print(set([1, 2, 2, 3]))
print(list({10, 20, 30}))

# string ↔ list
print(list("hello"))
print("".join(['h','i']))

#string <-> tuple
print(tuple("hello"))
print("".join(('h','i')))

# 3. Chained Casting

x="10.5"
print(x, type(x))
y=float(x)
print(y,type(y))
z=int(y)
print(z,type(z))

```

```python
data = ["25", "30", "40"]
cleaned = [int(x) for x in data]
print(cleaned)
Total_Sum = sum([int(x) for x in data])
print(Total_Sum)
Average = sum([int(x) for x in data])/len(data)
print(round(Average,2))

# Handling Missing data

data = ['10','20','NA','30']
# scenario : Replace missing data with 0 and then cast all elemnets to int
new_data= [int(x) if x!='NA' else 0 for x in data]
print(new_data)

# scenario : Replace missing data with 100
new_data= [100 if x=='NA' else int(x) for x in data]
print(new_data)

# Converting to boolean

data = ["yes","no","Yes","No"]
bool_data = [True if x.lower() == "yes" else False for x in data]
print(bool_data)

# Converting dates

from datetime import datetime

data = ["2025-08-16", "2025-08-15"]
cleaned = [datetime.strptime(x, "%Y-%m-%d") for x in data]
print(cleaned[0],cleaned[1])

# Cleaning Numeric data

"""
1. When we have to perform cleanup on numeric data then we have to import Pandas library
2. In pandas, we use "astype()" for type conversion
3. It returns results in index format
"""

import pandas as pd

df = pd.DataFrame({
    "age": ["25", "30", "NA", "40"],
    "salary": ["50000.50", "60000.00", "55000.25", "70000.00"]
})

# Convert age to integer, replacing "NA" with 0
df["age"] = df["age"].replace("NA", 0).astype(int)

# Convert salary to float
df["salary"] = df["salary"].astype(float)

print(df)
```

```python
# String

"""
1. Strings are immutable means once created, you cannot change individual characters.
2. Strings in Python are objects of class str.
3. positive index = from start and negative index = from end
4. Slicing format = [start:end:step]
"""

word = "Python"
print(word[::1])
print(word[::2])
print(word[::-1])
print(word[::-2])

# Membership

word = "python3"
print("py"in word)
print("py " in word)

word = "   Hello World  "
print(word)
print(word.strip()) # strip removes the whitespaces from the begining and end of the word
print(word.split())
```
